```python
#########################################################
## Task1
#########################################################

import numpy as np
from sklearn.model_selection import KFold
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD

# read the data file
def dataSet(file_path):
    with open(file_path) as file_in:
        data_set = []
        for line in file_in.readlines():
            line = line.strip().split()
            data_set.append(list(map(float,line)))
    return data_set  
```

```python
def dataOneAndFive(file_path):
    data_set = dataSet(file_path)
    data_one_and_five = []
    for figure in data_set:
        if 1.0 == figure[0] or 5.0 == figure[0]:
            data_one_and_five.append(figure)

    figure = np.array(data_one_and_five)[:,1:]
    label = np.array(data_one_and_five)[:,0]

    return figure,label

#create and process model
def modelProcess(input_dim,lr_schedule, figure, label):
    model = Sequential()
    model.add(Dense(input_dim, input_shape=(2,), activation='sigmoid', kernel_initializer='lecun_uniform'))
    model.add(Dense(1, activation='sigmoid', kernel_initializer='lecun_uniform'))
    model.compile(optimizer=SGD(learning_rate=lr_schedule), loss='binary_crossentropy', metrics=['accuracy'])
    
    model.fit(figure, label, epochs=100)
    return model

def trainModel(file_path):
    X, Y = dataOneAndFive(file_path)
    
    # 3-fold cross-validation
    k_fold = KFold(n_splits=3)
    for train, test in k_fold.split(X, Y):
        model1 = modelProcess(2, 0.01,X[train], Y[train])
        model2 = modelProcess(5, 0.01, X[train], Y[train])
        model3 = modelProcess(10, 0.01, X[train], Y[train])
        
        # evaluate the model
        # for model1
        in_samp_err1, in_samp_acc1 = model1.evaluate(X[train], Y[train])
        print("when input dimension = 2")
        print("the in-sample error is " + str(in_samp_err1))
        test_set_error1, test_set_acc1 = model1.evaluate(X[test], Y[test])
        print("the test-set error is " + str(test_set_error1))
        print("===========================================================")
        
        least_err = test_set_error1
        model1.save("q1_model1")
        
        # for model2
        in_samp_err2, in_samp_acc2 = model2.evaluate(X[train], Y[train])
        print("when input dimension = 5")
        print("the in-sample error is " + str(in_samp_err2))
        test_set_error2, test_set_acc2 = model2.evaluate(X[test], Y[test])
        print("the test-set error is " + str(test_set_error2))
        print("===========================================================")
        
        if test_set_acc2 <= least_err:
            least_err = test_set_error2
        model2.save("q1_model2")
            
        # for model3
        in_samp_err3, in_samp_acc3 = model3.evaluate(X[train], Y[train])
        print("when input dimension = 10")
        print("the in-sample error is " + str(in_samp_err3))
        test_set_error3, test_set_acc3 = model3.evaluate(X[test], Y[test])
        print("the test-set error is " + str(test_set_error3))
        
        if test_set_acc3 <= least_err:
            least_err = test_set_error3
        model3.save("q1_model3")
    
    if test_set_error1 == least_err:
        print("the best model is model1 with test set error equals to " + str(least_err))
    elif test_set_error2 == least_err:
        print("the best model is model2 with test set error equals to " + str(least_err))
    elif test_set_error3 == least_err:
        print("the best model is model3 with test set error equals to " + str(least_err))
trainModel('train.txt')
```

From comparing the test set error, I got that the model1 which have 2 dimensions is the best.

```python
from keras.utils import plot_model
from keras.models import load_model

def testData(file_path):
    X, Y = dataOneAndFive(file_path)

    # model1
    model1 = load_model('q1_model1')
    model1.summary()
    model1_img = 'model_1.png'
    plot_model(model1, to_file = model1_img, show_shapes=True)
    loss1, accuracy1 = model1.evaluate(X,Y)
    print("for model1, the loss equals to " + str(loss1) + " and the accuracy equals to " + str(accuracy1))
    
    # model2
    model2 = load_model('q1_model2')
    model2.summary()
    model2_img = 'model_2.png'
    plot_model(model2, to_file = model2_img, show_shapes=True)
    loss2, accuracy2 = model2.evaluate(X,Y)
    print("for model2, the loss equals to " + str(loss2) + " and the accuracy equals to " + str(accuracy2))
    
    # model3
    model3 = load_model('q1_model3')
    model3.summary()
    model3_img = 'model_3.png'
    plot_model(model3, to_file = model3_img, show_shapes=True)
    loss3, accuracy3 = model3.evaluate(X,Y)
    print("for model3, the loss equals to " + str(loss3) + " and the accuracy equals to " + str(accuracy3))

testData("test.txt")
```
